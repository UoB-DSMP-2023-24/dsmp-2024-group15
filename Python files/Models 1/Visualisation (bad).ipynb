```python
import os
import pandas as pd
import matplotlib.pyplot as plt

#Current working directory
script_dir = os.getcwd()

#Data file path, relative to CWD
file_path = os.path.join(script_dir, '..', '..', 'Prompt', 'Data', 'LOB sorted')

data = pd.read_csv(file_path)
data = data.drop(columns=['Unnamed: 0'])

print(data.head(10))

bid_condition = (data['Type'] == 'bid')
ask_condition = (data['Type'] == 'ask')

bids = data[(data['Type'] == 'bid')]
asks = data[(data['Type'] == 'ask')]

```

```python
plt.plot(bids["Time"],bids["Price"])
plt.title("Bids over time")
plt.ylabel("Bid Price")
plt.xlabel("Time")
plt.show()
```

```python
plt.plot(asks["Time"],asks["Price"])
plt.title("Asks over time")
plt.ylabel("Ask Price")
plt.xlabel("Time")
plt.show()
```
