```python
import os

# Get the path address of the csv file in the folder
def get_csv_urls(folder_path):
    csv_urls = []
    file_names = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".csv"):
                csv_urls.append(os.path.join(root, file))
                file_names.append(file)
    return csv_urls,file_names


# Iterate over the csv path list, read the data and splice it
def get_csv_data(csv_urls, file_names):
    for csv_url in csv_urls:
        # Read csv file
        # print(csv_url)
        df = pd.read_csv(csv_url, encoding='utf-8',low_memory=False,header=None)
        # Create a new date column with the content file_name
        df['date'] = file_names[csv_urls.index(csv_url)]

        # Splicing csv data
        try:
            csv_data = pd.concat([csv_data, df], axis=0)
        except:
            csv_data = df

    return csv_data

# Data file address
folder_path = r'E:\vsc_workspace\ml_learn\ml_02_03\Tapes'

# Get csv path
csv_urls,file_names = get_csv_urls(folder_path)
print(csv_urls)
print(file_names)

# Get csv data
csv_data = get_csv_data(csv_urls, file_names)
print(len(csv_data))
csv_data.head()
```

```python
# Clean date, extracting it from a string like file_name.
csv_data['date'] = csv_data['date'].str.extract('(\d{4}-\d{2}-\d{2})')
csv_data['date'] = pd.to_datetime(csv_data['date'])

# Deleting null values
csv_data = csv_data.dropna()
# Reset Index
csv_data = csv_data.reset_index(drop=True)

# Export csv
csv_data.to_csv('tapes.csv', encoding='utf-8', index=False)
csv_data.head()
```
